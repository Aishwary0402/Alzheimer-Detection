The columns which have 2 unique values we will not include that column in the boxplot because the box plot of the value(0 and 1) will be a point where the outlier will be 1, so there is no point of plotting the boxplot of the features which have only 2 values.

```python
# Box plot for all the features for checking the outliers
fil_cols = [col for col in df.columns if df[col].nunique() > 2]  

for col in df[fil_cols].columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    
    plt.title(f'Boxplot of{col}')
    plt.ylabel(col)
    plt.show()
```

So we can see from the box plot that there are no outliers in the columns so we dont have to remove any datapoints or since we dont have outliers, so will we not be using the outliers detection techniques IQR and Z-score.

```python
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.distplot(x=df[col])
    plt.title(f'Distplot of {col}')
    plt.ylabel(col)
    plt.show()
```

# Feature Selection

```python
#We will draw the correlation heatmap (only on the top features
corr_matrix = df.corr().abs()
plt.figure(figsize=(12, 8))  # Adjust size for better visibility
sns.heatmap(corr_matrix.nlargest(27,'Diagnosis')['Diagnosis'].to_frame(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Feature Correlations")
plt.show()
```

```python
top_27_features = corr_matrix.nlargest(27, 'Diagnosis')['Diagnosis'].index.tolist()
print(top_27_features)
```

```python
total_variance=df.var().sum()
selected_variance=df[top_27_features].var().sum()
```

```python

variance_explained = (selected_variance / total_variance) * 100
print(f"Total Variance Captured by Top 27 Features: {variance_explained:.2f}%")
```

 These are the most important features of our dataset that will help in the Prediction

# Normalization

```python
#Normalizing the features
scaler=MinMaxScaler()
x_sample=scaler.fit_transform(df)
print(x_sample)
```

# Principle Component Analysis(PCA)

Now since we have soo many features so we will reduce the features bty applying the Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
#Applying PCA with 4 components
pca=PCA(n_components=0.95)
x_pca=pca.fit_transform(x_sample)
```

```python
explained_varience=pca.explained_variance_ratio_
print(f"Explained Varience: {explained_varience}")
print(f"Total Varience Captured: {sum(explained_varience)*100:.2f}%")
```

So here we can see that after applying the pca also to reach 95% of varience we have too use at least 30 features so we will not apply PCA for the dataset we will do the rest things with our original dataset without applying the PCA

# Splitting the Dataset into Train and Test

Before splitting the dataset, we have to convert the normalized dataset into a dataframe because after normalizing, it has become into a numpy array, so will convert it into a dataframe 

```python
dff=pd.DataFrame(x_sample)
```

```python
dff.head()
```

```python
X=dff.drop(columns=32)
```

```python
Y=df["Diagnosis"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

# Model Fitting

```python
# Import ML models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Support Vector Machine": SVC(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}
```

# Model Accuracy

```python
from sklearn.metrics import accuracy_score, classification_report

# Training models and evaluate performance
results_sampled = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred,output_dict=True)
    results_sampled[name] = {"Accuracy": accuracy, "Report": report}
    # Converting results into a DataFrame
    results_df = pd.DataFrame({
    model: {
    "Accuracy": metrics["Accuracy"],
    "Precision": metrics["Report"]["weighted avg"]["precision"],
    "Recall": metrics["Report"]["weighted avg"]["recall"],
    "F1-Score": metrics["Report"]["weighted avg"]["f1-score"]
    }
    for model, metrics in results_sampled.items()
    })
# Transposing to get models as column headers
results_df = results_df.T
# Displaying results
print(results_df)
```

# Visualization of Accuarcy Scores

```python
# Extracting accuracy scores for visualization
model_names = list(results_sampled.keys())
accuracies = [results_sampled[model]["Accuracy"] for model in model_names]
# Plotting accuracy comparison
plt.figure(figsize=(8, 5))
sns.barplot(x=model_names, y=accuracies, palette="viridis")
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
# Selecting the first four models
selected_models = list(models.keys())[:4]
# Creating subplots
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
for i, name in enumerate(selected_models):
    model = models[name]
    y_pred = model.predict(X_test) # Predictions
    cm = confusion_matrix(y_test, y_pred)
    # Heatmap
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",xticklabels=set(y_test),
    yticklabels=set(y_test), ax=axes[i])
    axes[i].set_title(f'Confusion Matrix - {name}')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')

# Adjusting layout and showing the plot
plt.tight_layout()
plt.show()
```
